```python
from sklearn import linear_model
```

```python
# input => 2D  |   #output => 1D

hb = df[['Hemoglobin (g/dL)']]
hb
```

```python
pl = df['Platelets (x10^9/L)']
```

```python
pl.head()
```

```python
# create linear regression object
reg = linear_model.LinearRegression()
```

```python
reg.fit(hb,pl)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
# y = mx + b     |  b=> intercept    |   m => coef
pl = reg.coef_[0]*14.5 + reg.intercept_
```

```python
pl
```

```python
reg.predict([[11.2]])
```
